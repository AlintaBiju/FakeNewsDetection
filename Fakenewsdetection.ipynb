```python
def wordopt(text):
  text = text.lower()
  text = re.sub(r'https?://\S+','',text)
  text = re.sub(r'@\S+',' ',text)
  text = re.sub(r'[^\w\s]',' ',text)
  text = re.sub(r'\d','',text)
  text = re.sub(r'\n','',text)
  return text
```

```python
news['text'] = news['text'].apply(wordopt)
```
